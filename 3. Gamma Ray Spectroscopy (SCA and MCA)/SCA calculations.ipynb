**SCA**

Performing a Gaussian fitting to find the FWHM (Full Width at Half Maximum) on the data points for different operating volatges and comparing their resolution to find the optimal one.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from prettytable import PrettyTable
```

```python
def Gaussian(x, A, x0, sigma):
    return A * np.exp(-(x - x0)**2 / (2 * sigma**2))

#Data points (Operating volatge = 450V)
x = np.array([2.4,2.5,2.6,2.7,2.8,2.9,3.0,3.1,3.2,3.3,3.4,3.5])
y = np.array([625,601,928,1659,3700,6406,7536,5102,2222,717,218,118])


popt, pcov = curve_fit(Gaussian, x, y,
                       p0=[y.max(), x[np.argmax(y)], 0.1])

A, x0, sigma = popt
FWHM = 2*np.sqrt(2*np.log(2))*sigma
R = (FWHM / x0) * 100

half_max = A / 2
x_left = x0 - FWHM / 2
x_right = x0 + FWHM / 2

print(f"Resolution: {R:.2f} %")
print(f"FWHM: {FWHM:.4f} V")
print(f"Mean Voltage (x0): {x0:.2f} V")
print(f"Standard Deviation (sigma): {sigma:.4f} V")

# Reduced chi-squared
sigma_y = np.sqrt(y)
residuals = y - Gaussian(x, *popt)
chi2 = np.sum((residuals / sigma_y)**2)
ndof = len(x) - len(popt)
chi2_red = chi2 / ndof
print(f"Reduced chi-squared: {chi2_red:.3f}")

plt.scatter(x, y, label='Data', color='red')
x_fit = np.linspace(x.min(), x.max(), 1000)
y_fit = Gaussian(x_fit, *popt)
plt.plot(x_fit, y_fit, label='Gaussian Fit', color='blue')
plt.axhline(half_max, linestyle='--', linewidth=1.5, label='Half Maximum')
plt.axvline(x_left, linestyle='--', linewidth=1.5, label='FWHM Left')
plt.axvline(x_right, linestyle='--', linewidth=1.5, label='FWHM Right')
plt.xlim(0,4)
plt.xlabel('LLD Voltage (V)')
plt.ylabel('Counts')
plt.title('SCA Counts vs LLD Voltage Gaussian Fit Operating volatge = 450V')
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.array([662.399,355.829,1171.402,1335.363])
y = np.array([298.9051,170.7242,516.2521,587.5273])
n = len(x)

# --- Data Table ---
table_data = PrettyTable(["i", "x", "y", "x", "xy"])
for i in range(n):
    table_data.add_row([i+1, round(x[i], 3), round(y[i], 3), round(x[i]**2, 3), round(x[i]*y[i], 3)])
table_data.add_row(["", round(np.sum(x), 3), round(np.sum(y), 3), round(np.sum(x**2), 3), round(np.sum(x*y), 3)])
table_data.title = "Input Data"
print(table_data, "\n")

# --- Least Squares Calculation ---
Sx = np.sum(x)
Sy = np.sum(y)
Sxx = np.sum(x**2)
Sxy = np.sum(x*y)

delta = n*Sxx - Sx**2
a1 = (n*Sxy - Sx*Sy) / delta  # Slope
a0 = (Sxx*Sy - Sx*Sxy) / delta  # Intercept

# --- Error Analysis ---
y_fit = a0 + a1*x
residuals = y - y_fit
sigma_y = np.sqrt(np.sum(residuals**2) / (n-2))
sigma_a1 = sigma_y * np.sqrt(n / delta)
sigma_a0 = sigma_y * np.sqrt(Sxx / delta)

# --- Results Table ---
results = PrettyTable(["Parameter", "Value", "Uncertainty"])
results.title = "Fit Results & Statistical Errors"
results.add_row(["Slope (a)", f"{a1:.4f}", f"{sigma_a1:.4f}"])
results.add_row(["Intercept (a)", f"{a0:.4f}", f"{sigma_a0:.4f}"])
results.add_row(["_y (Std. Error)", f"{sigma_y:.4f}", ""])
results.add_row([" (delta)", f"{delta:.4f}", ""])
print(results, "\n")

# --- Fitted Values Table ---
fit_table = PrettyTable(["x", "y (measured)", "y (fit)"])
fit_table.title = "Measured vs Fitted Values"
for xi, yi, yfi in zip(x, y, y_fit):
    fit_table.add_row([round(xi, 3), round(yi, 3), round(yfi, 3)])
print(fit_table, "\n")

# --- Plotting ---
plt.figure(figsize=(7,5))
plt.scatter(x, y, color='navy', label='Measured Data', s=40)
plt.plot(x, y_fit, color='crimson', linewidth=2,
         label=f'Least Squares Fit:\n$y = ({a1:.5f} \pm {sigma_a1:.5f})x + ({a0:.5f} \pm {sigma_a0:.5f})$')
plt.xlabel("Magnetic field (Gauss)", fontsize=12)
plt.ylabel("Corrected Hall Voltage (mV)", fontsize=12)
plt.title("Hall Voltage vs Magnetic field (n-type Si)", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

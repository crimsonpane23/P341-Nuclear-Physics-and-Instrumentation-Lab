**SCA**

Performing a Gaussian fitting to find the FWHM (Full Width at Half Maximum) on the data points for different operating volatges and comparing their resolution to find the optimal one.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
import numpy as np
from scipy.optimize import curve_fit

def Gaussian(x, A, x0, sigma):
    return A * np.exp(-(x - x0)**2 / (2 * sigma**2))

#Data points (Operating volatge = 450V)
x = np.array([2.4,2.5,2.6,2.7,2.8,2.9,3.0,3.1,3.2,3.3,3.4,3.5])
y = np.array([625,601,928,1659,3700,6406,7536,5102,2222,717,218,118])


popt, pcov = curve_fit(Gaussian, x, y,
                       p0=[y.max(), x[np.argmax(y)], 0.1])

A, x0, sigma = popt
FWHM = 2*np.sqrt(2*np.log(2))*sigma
R = (FWHM / x0) * 100

half_max = A / 2
x_left = x0 - FWHM / 2
x_right = x0 + FWHM / 2

print(f"Resolution: {R:.2f} %")
print(f"FWHM: {FWHM:.4f} V")
print(f"Mean Voltage (x0): {x0:.2f} V")
print(f"Standard Deviation (sigma): {sigma:.4f} V")

# Reduced chi-squared
sigma_y = np.sqrt(y)
residuals = y - Gaussian(x, *popt)
chi2 = np.sum((residuals / sigma_y)**2)
ndof = len(x) - len(popt)
chi2_red = chi2 / ndof
print(f"Reduced chi-squared: {chi2_red:.3f}")

plt.scatter(x, y, label='Data', color='red')
x_fit = np.linspace(x.min(), x.max(), 1000)
y_fit = Gaussian(x_fit, *popt)
plt.plot(x_fit, y_fit, label='Gaussian Fit', color='blue')
plt.axhline(half_max, linestyle='--', linewidth=1.5, label='Half Maximum')
plt.axvline(x_left, linestyle='--', linewidth=1.5, label='FWHM Left')
plt.axvline(x_right, linestyle='--', linewidth=1.5, label='FWHM Right')
plt.xlim(0,4)
plt.xlabel('Voltage (V)')
plt.ylabel('Counts')
plt.title('SCA Counts vs Voltage with Gaussian Fit')
plt.legend()
plt.grid()
plt.show()
```
